# Lotka-Volterra fits

Lotka-Volterra is a system of differential equation used to model the dynamics of biological system as for example the predator-prey interaction. This set of equation state that the population change according to these equation:
$$
\dfrac{dx}{dt} = (b - p y)x \\
\dfrac{dy}{dt} = (rx - d)y
$$

with:
* x the number of prey
* y the number of predator 
* $\dfrac{dx}{dt}$ and $\dfrac{dy}{dt}$ the variation rate over time of the populations
* $b, p, r, d$ positive real parameters describing the dynamics

Now as requested from the exercise I'll proceed finding the best values that fits the Snowshoe hare, Canada Lynx populations.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
import sklearn
from sklearn.metrics import mean_squared_error
import pysindy as ps
from scipy import integrate
from scipy import optimize 
from scipy import interpolate



years = np.linspace(1845, 1903, 30)
snowshoe_hare = np.array([20,20,52,83,64,68,83,12,36,150,110,60,7,10,70,100,92,70,10,11,137,137,18,22,52,83,18,10,9,65]);
canada_lynx = np.array([32,50,12,10,13,36,15,12,6,6,65,70,40,9,20,34,45,40,15,15,60,80,26,18,37,50,35,12,12,25]);

fig = plt.figure(1,figsize=(8, 6), dpi=80)
line1 = plt.plot(years, snowshoe_hare, label='Snowshoe hare')
line2 = plt.plot(years, canada_lynx, label='Canada Lynx')
plt.xlabel('years')
plt.ylabel('population [x1000]')
plt.title('Population')
plt.legend(loc='upper right')
plt.show()
```

```python
time = np.array(years[:] - years[0])
dt = int(years[1] - years[0])

def lotkavolterra(t, z, b, p, r, d):

    x, y = z

    return [b*x - p*x*y, -d*y + r*x*y]


x0 = [20,32]
b_ini = 1.5; p_ini = 1;
r_ini = 1; d_ini = 3;
#b_ini = 0; p_ini = 0;
#r_ini = 0; d_ini = 0;

init_argument = (b_ini, p_ini, r_ini, d_ini)

sol = integrate.solve_ivp(lotkavolterra, [time[0], time[len(time)-1]], x0, args=init_argument,
                                                          dense_output=True, method = 'RK45', t_eval=time)
```

Once the Lotka-Volterra function and is integration is tested, is it possible to proceed with the parameter optimization.

```python
def optimization_fun(x):
    arguments=(x[0], x[1], x[2], x[3])
    sol = integrate.solve_ivp(lotkavolterra, [time[0], time[len(time)-1]], x0, args=arguments,
                                                          dense_output=True, method = 'RK45', t_eval=time)
    
    res = np.sum((snowshoe_hare - sol.y[0,:])**2) + np.sum((canada_lynx - sol.y[1,:])**2)
    return res



res = optimize.fmin(optimization_fun, init_argument)#, method='Powell', options={'xatol': 1e-8, 'disp': True})
```

```python
print(res)
sol = integrate.solve_ivp(lotkavolterra, [time[0], time[len(time)-1]], x0, args=res,
                                                          dense_output=True, method = 'RK45', t_eval=time)

fig = plt.figure(2,figsize=(8, 6), dpi=80)
line1 = plt.plot(years, sol.y[0,:], label='Snowshoe hare LV')
line2 = plt.plot(years, sol.y[1,:], label='Canada Lynx LV')
line3 = plt.plot(years, snowshoe_hare, '--', linewidth='2', color='tab:blue', alpha=.5, label='Snowshoe hare')
line4 = plt.plot(years, canada_lynx, '--', linewidth='2', color='tab:orange', alpha=.5, label='Canada Lynx')

plt.xlabel('years')
plt.ylabel('population [x1000]')
plt.title('Population')
plt.legend(loc='upper right')
plt.show()
```

Trying augmenting the number of the time series. I decide to use a cubic interpolation between the data.

```python
hd_dt = 0.1
hd_time = np.arange(time[0], time[len(time)-1] + hd_dt, hd_dt)
hd_years = years[0] + hd_time

interp_snowshoe_hare = interpolate.interp1d(time, snowshoe_hare, kind='cubic')
interp_canada_lynx = interpolate.interp1d(time, canada_lynx, kind='cubic')

hd_snowshoe_hare = interp_snowshoe_hare(hd_time)
hd_canada_lynx = interp_canada_lynx(hd_time)


fig = plt.figure(3,figsize=(8, 6), dpi=80)
line1 = plt.plot(hd_years, hd_snowshoe_hare, label='Snowshoe hare')
line2 = plt.plot(hd_years, hd_canada_lynx, label='Canada Lynx')
line3 = plt.plot(years, snowshoe_hare, '-v', linewidth='2', color='tab:blue', alpha=.5)
line4 = plt.plot(years, canada_lynx, '-v', linewidth='2', color='tab:orange', alpha=.5)

plt.xlabel('years')
plt.ylabel('population [x1000]')
plt.title('Population')
plt.legend(loc='upper right')
plt.show()
```
